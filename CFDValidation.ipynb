```python

# ============================================================
# Reduced Finite-Difference Solver for Closure Validation
#
# Includes:
#   1) Discovered algebraic turbulence closure
#   2) Independent mixing-length solution with wall correction
#
# Discovered closure unknowns:
#   phi_L_interface
#   phi_g_interface
#   nut_interface_L_star
#   nut_interface_g_star
#
# Discovered closure constraints:
#   meanL = 1
#   meanG = 1
#   u_i,L = u_i,g
#   tau_L = tau_g
#
# Mixing-length unknowns:
#   phi_L_interface
#   phi_g_interface
#
# Mixing-length constraints:
#   u_i,L = u_i,g
#   tau_L = tau_g
#
# Important:
#   For mixing-length, meanL and meanG are reported,
#   but they are NOT forced to be 1.
#
# Mixing-length distance definition:
#   Liquid:
#       d = d_wall
#
#   Gas:
#       d = min(d_wall, d_interface)
# ============================================================

from google.colab import drive
drive.mount("/content/drive")

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.optimize import root, root_scalar
from scipy.linalg import solve_banded


# ============================================================
# INPUT DATA
# ============================================================

Y_I_L = np.array([
    -0.38599138, -0.50685566, -0.54353263, -0.62025884,
    -0.19339468, -0.36639731, -0.39461579, -0.45593355
])

Y_I_g = np.array([
    -0.21100252, -0.24296018, -0.25118830, -0.26666674,
    -0.13647076, -0.20497163, -0.21357199, -0.23048830
])

Vel_ave_L = np.array([
    0.13559322, 0.15564202, 0.16194332, 0.17543860,
    0.15915119, 0.19867550, 0.20547945, 0.22058824
])

Vel_ave_g = np.array([
    0.78156028, 1.76043069, 2.04249668, 2.98704663,
    0.88443018, 1.87392550, 2.17231638, 3.16758242
])

Re_L = np.array([
    3450.35355310, 3604.56651513, 3650.38144777, 3744.77563182,
    4791.70675035, 5137.32386263, 5192.28471770, 5310.30353933
])

Re_g = np.array([
    2444.18629244, 5714.13581516, 6692.97239000, 9963.29544459,
    2535.15810329, 5819.15232740, 6813.96014038, 10133.83525732
])

dPdX_L = np.array([
    -0.00067507, -0.00187867, -0.00228591, -0.00348701,
    -0.00079928, -0.00172978, -0.00205301, -0.00303126
])

dPdX_g = np.array([
    -0.03091919, -0.02548305, -0.02586579, -0.02327339,
    -0.03050943, -0.02891136, -0.02823237, -0.02418915
])

R_L = np.array([
    1.17682855, 1.29304204, 1.32850627, 1.40293698,
    0.99462298, 1.15810032, 1.18508287, 1.24394684
])

R_g = np.array([
    0.64331436, 0.61981694, 0.61395621, 0.60316212,
    0.70186498, 0.64786969, 0.64138464, 0.62885302
])

Dh_L = np.array([
    0.02549224, 0.02320110, 0.02258175, 0.02138371,
    0.03016218, 0.02590449, 0.02531469, 0.02411679
])

Dh_g = np.array([
    0.04663350, 0.04840139, 0.04886342, 0.04973787,
    0.04274326, 0.04630561, 0.04677380, 0.04770590
])

rho_L = 998.2
rho_g = 1.2

# Used only as initial guesses for the discovered-closure solver.
nut_interface_L_star_guess = np.array([
    0.01388679, 0.09201774, 0.11511446, 0.09027585,
    0.07092999, 0.01215777, 0.02467957, 0.01381721
])

nut_interface_g_star_guess = np.array([
    0.00163215, 0.00188662, 0.00219779, 0.00236930,
    0.00123606, 0.00145128, 0.00174341, 0.00202939
])

# Physical interface location:
# y_phys = Y_L * Dh_L
y_I = Y_I_L * Dh_L


# ============================================================
# SETTINGS
# ============================================================

N_L = 400
N_G = 400

PLOT_RESULTS = True
PLOT_CFD_DOTS = True
PLOT_MOMENTUM_RESIDUALS = False
PRINT_SUMMARY = True
RUN_GRID_REFINEMENT = True

base_path = "/content/drive/MyDrive/CFDdata"

# Shear convention:
#   "difference" -> tauL - tauG = 0
#   "sum"        -> tauL + tauG = 0
#
# Use the same convention in the paper, PINN loss, and FD solver.
SHEAR_CONVENTION = "difference"

EPS = 1e-14


# ============================================================
# MIXING-LENGTH SETTINGS
# ============================================================

KAPPA = 0.41
A_PLUS = 26.0
C_ELL_CAP = 0.09

MIX_MAX_ITER = 200
MIX_TOL = 1e-8
MIX_RELAX = 0.5


# ============================================================
# HELPER FUNCTIONS
# ============================================================

def trapz_mean(U, Y):
    """
    Mean value over a 1D interval:
        mean(U) = integral U dY / interval length
    """
    return np.trapz(U, Y) / (Y[-1] - Y[0])


def dUdy_first(U, Y):
    """
    First-order derivative at the left boundary.
    """
    h = Y[1] - Y[0]
    return (U[1] - U[0]) / h


def dUdy_last(U, Y):
    """
    First-order derivative at the right boundary.
    """
    h = Y[1] - Y[0]
    return (U[-1] - U[-2]) / h


def gradient_full(U, Y):
    """
    Full-grid dU/dY.
    """
    dU = np.zeros_like(U)

    dU[1:-1] = (U[2:] - U[:-2]) / (Y[2:] - Y[:-2])
    dU[0] = (U[1] - U[0]) / (Y[1] - Y[0])
    dU[-1] = (U[-1] - U[-2]) / (Y[-1] - Y[-2])

    return dU


def fixed_velocity_scale(VL, VG):
    return max(0.5 * (abs(VL) + abs(VG)), 1e-12)


def fixed_shear_scale(VL, VG):
    return max(0.015 * 0.5 * rho_g * (VG - VL)**2, 1e-12)


def interface_shear_residual(tauL, tauG):
    if SHEAR_CONVENTION == "difference":
        return tauL - tauG
    elif SHEAR_CONVENTION == "sum":
        return tauL + tauG
    else:
        raise ValueError("SHEAR_CONVENTION must be 'difference' or 'sum'.")


# ============================================================
# DISCOVERED TURBULENCE CLOSURES
# ============================================================

def liquid_nut_star(YL, RL, YIL, ReL_val, nut_i_L_star):
    """
    Liquid closure:
        nut_L / nut_L_interface
        =
        eta^2 [1 + A (eta (1 - eta))^2]

    where:
        eta = d_wall / delta
        A = 0.011744 Re_L - 38.672110
    """
    delta = YIL + RL
    d_wall = YL + RL
    eta = np.clip(d_wall / delta, 0.0, 1.0)

    A = 0.011744 * ReL_val - 38.672110

    shape = eta**2 * (1.0 + A * (eta * (1.0 - eta))**2)

    return np.maximum(nut_i_L_star * shape, 0.0)


def gas_nut_star(YG, RG, YIG, ReL_val, ReG_val, nut_i_g_star):
    """
    Gas closure:
        nut_g / nut_peak = (1 - alpha) b_wall + alpha b_interface
    """
    delta = RG - YIG
    d_wall = RG - YG

    ratio_Re = ReG_val / ReL_val

    r_i = 0.207886 * ratio_Re - 0.056330
    r_i = np.clip(r_i, 1e-8, 0.999999)

    nut_peak_star = nut_i_g_star / r_i

    X = (1.0 - r_i) * ratio_Re

    eta_peak = -0.268621 * X + 0.645132
    eta_peak = np.clip(eta_peak, 1e-6, 0.999999)

    d_peak = eta_peak * delta

    c1 = 2.313850 * X - 3.359563
    c2 = 12.055194 * X - 6.164560

    b_wall = (d_wall / (d_peak + EPS))**2 * (
        1.0 + c1 * d_wall * (d_peak - d_wall) / (d_peak**2 + EPS)
    )

    denom = delta - d_peak

    if abs(denom) < EPS:
        denom = EPS if denom >= 0.0 else -EPS

    b_interface = r_i + (1.0 - r_i) * ((delta - d_wall) / denom)**2 * (
        1.0 + c2 * ((d_peak - d_wall) * (delta - d_wall)) / (denom**2 + EPS)
    )

    alpha = d_wall**2 / (
        d_wall**2 + d_peak**2 * ((delta - d_wall) / denom)**2 + EPS
    )

    shape_peak = (1.0 - alpha) * b_wall + alpha * b_interface
    shape_peak = np.maximum(shape_peak, 0.0)

    nut_star = np.maximum(nut_peak_star * shape_peak, 0.0)

    return {
        "nut_star": nut_star,
        "nut_peak_star": nut_peak_star,
        "r_i": r_i,
        "eta_peak": eta_peak,
        "c1": c1,
        "c2": c2,
        "X": X,
    }


# ============================================================
# FINITE-DIFFERENCE SOLVER
# ============================================================

def solve_ode_fd(Y, mu_eff, rhs_const,
                 bc_left_type, bc_left_val,
                 bc_right_type, bc_right_val):
    """
    Solve:
        d/dY( mu_eff dU/dY ) = rhs_const

    Boundary conditions:
        Dirichlet: U = value
        Neumann:  mu_eff dU/dY = value
    """
    n = len(Y)
    h = Y[1] - Y[0]

    lower = np.zeros(n)
    diag = np.zeros(n)
    upper = np.zeros(n)
    b = np.zeros(n)

    mu_face = 0.5 * (mu_eff[:-1] + mu_eff[1:])

    # Left boundary
    if bc_left_type == "dirichlet":
        diag[0] = 1.0
        b[0] = bc_left_val

    elif bc_left_type == "neumann":
        diag[0] = -mu_eff[0] / h
        upper[0] = mu_eff[0] / h
        b[0] = bc_left_val

    else:
        raise ValueError("Unknown left boundary condition type.")

    # Interior nodes
    for j in range(1, n - 1):
        mu_m = mu_face[j - 1]
        mu_p = mu_face[j]

        lower[j] = mu_m / h**2
        diag[j] = -(mu_m + mu_p) / h**2
        upper[j] = mu_p / h**2
        b[j] = rhs_const

    # Right boundary
    if bc_right_type == "dirichlet":
        diag[-1] = 1.0
        b[-1] = bc_right_val

    elif bc_right_type == "neumann":
        lower[-1] = -mu_eff[-1] / h
        diag[-1] = mu_eff[-1] / h
        b[-1] = bc_right_val

    else:
        raise ValueError("Unknown right boundary condition type.")

    ab = np.zeros((3, n))
    ab[0, 1:] = upper[:-1]
    ab[1, :] = diag
    ab[2, :-1] = lower[1:]

    return solve_banded((1, 1), ab, b)


def momentum_residual_fd(Y, U, mu_eff, rhs_const):
    """
    Compute residual:
        residual = rhs_const - d/dY(mu_eff dU/dY)
    """
    n = len(Y)
    h = Y[1] - Y[0]

    residual_full = np.full(n, np.nan)

    mu_face = 0.5 * (mu_eff[:-1] + mu_eff[1:])

    for j in range(1, n - 1):
        mu_m = mu_face[j - 1]
        mu_p = mu_face[j]

        diffusion_term = (
            mu_p * (U[j + 1] - U[j])
            -
            mu_m * (U[j] - U[j - 1])
        ) / h**2

        residual_full[j] = rhs_const - diffusion_term

    residual_interior = residual_full[1:-1]

    rms_abs = np.sqrt(np.mean(residual_interior**2))
    max_abs = np.max(np.abs(residual_interior))

    denom = max(abs(rhs_const), 1e-14)

    return {
        "residual_full": residual_full,
        "residual_interior": residual_interior,
        "rms_abs": rms_abs,
        "rms_rel": rms_abs / denom,
        "max_abs": max_abs,
        "max_rel": max_abs / denom,
    }


# ============================================================
# INDEPENDENT MEAN-FLOW SOLVE
# Used only to generate initial guesses.
# ============================================================

def solve_phase_with_mean_constraint(Y, mu_eff, rhs_const, wall_at_left=True):
    """
    Find interface flux phi such that mean(U) = 1.
    This is used only for initial guesses.
    """
    def residual(phi):
        if wall_at_left:
            U = solve_ode_fd(
                Y, mu_eff, rhs_const,
                bc_left_type="dirichlet", bc_left_val=0.0,
                bc_right_type="neumann", bc_right_val=phi
            )
        else:
            U = solve_ode_fd(
                Y, mu_eff, rhs_const,
                bc_left_type="neumann", bc_left_val=phi,
                bc_right_type="dirichlet", bc_right_val=0.0
            )

        return trapz_mean(U, Y) - 1.0

    phi_values = np.concatenate([
        -10.0**np.arange(-10, 6),
         10.0**np.arange(-10, 6)
    ])

    phi_values = np.unique(np.sort(phi_values))
    f_values = np.array([residual(phi) for phi in phi_values])

    bracket = None

    for k in range(len(phi_values) - 1):
        if (
            np.isfinite(f_values[k])
            and np.isfinite(f_values[k + 1])
            and f_values[k] * f_values[k + 1] < 0.0
        ):
            bracket = (phi_values[k], phi_values[k + 1])
            break

    if bracket is not None:
        sol = root_scalar(
            residual,
            bracket=bracket,
            method="brentq",
            maxiter=200
        )
    else:
        sol = root_scalar(
            residual,
            x0=-1e-3,
            x1=1e-3,
            method="secant",
            maxiter=200
        )

    if not sol.converged:
        raise RuntimeError("Initial independent mean-flow solve did not converge.")

    phi_star = sol.root

    if wall_at_left:
        U = solve_ode_fd(
            Y, mu_eff, rhs_const,
            bc_left_type="dirichlet", bc_left_val=0.0,
            bc_right_type="neumann", bc_right_val=phi_star
        )
    else:
        U = solve_ode_fd(
            Y, mu_eff, rhs_const,
            bc_left_type="neumann", bc_left_val=phi_star,
            bc_right_type="dirichlet", bc_right_val=0.0
        )

    return phi_star, U


# ============================================================
# MIXING-LENGTH MODEL
# ============================================================

def mixing_length_nut_star_from_U(
    Y, U,
    phase,
    R_val,
    YI_val,
    Re_val,
    Dh_val,
    V_val
):
    """
    Mixing-length eddy viscosity:

        nut = l_m^2 |du/dy|

    Nondimensional:

        nut_star = nut / (V Dh)
                 = (l_m / Dh)^2 |dU/dY|

    Distance definition:
        Liquid:
            d = d_wall

        Gas:
            d = min(d_wall, d_interface)
    """
    dU_dY = gradient_full(U, Y)

    if phase == "liquid":
        delta_nd = YI_val + R_val

        d_wall_nd = Y + R_val
        d_interface_nd = YI_val - Y

        # Required:
        d_mix_nd = d_wall_nd

        # Wall is at left boundary
        dU_wall = dUdy_first(U, Y)

    elif phase == "gas":
        delta_nd = R_val - YI_val

        d_wall_nd = R_val - Y
        d_interface_nd = Y - YI_val

        # Required:
        d_mix_nd = np.minimum(d_wall_nd, d_interface_nd)

        # Wall is at right boundary
        dU_wall = dUdy_last(U, Y)

    else:
        raise ValueError("phase must be 'liquid' or 'gas'.")

    d_wall_nd = np.maximum(d_wall_nd, 0.0)
    d_mix_nd = np.maximum(d_mix_nd, 0.0)

    d_wall_phys = d_wall_nd * Dh_val
    d_mix_phys = d_mix_nd * Dh_val
    delta_phys = delta_nd * Dh_val

    # Molecular kinematic viscosity:
    # Re = V Dh / nu
    nu_mol = V_val * Dh_val / Re_val

    # Wall friction velocity estimate:
    # u_tau = V sqrt(|(1/Re) dU/dY_wall|)
    u_tau = V_val * np.sqrt(
        max(abs((1.0 / Re_val) * dU_wall), 1e-16)
    )

    d_wall_plus = d_wall_phys * u_tau / max(nu_mol, 1e-16)

    # Van Driest damping
    damping = 1.0 - np.exp(-d_wall_plus / A_PLUS)

    # Mixing length without outer cap
    ell_phys = KAPPA * d_mix_phys * damping

    nut_star = (ell_phys / Dh_val)**2 * np.abs(dU_dY)

    nut_star = np.nan_to_num(nut_star, nan=0.0, posinf=0.0, neginf=0.0)
    nut_star = np.maximum(nut_star, 0.0)

    return nut_star


def solve_mixing_length_given_phi(
    Y,
    rhs_const,
    phi,
    phase,
    R_val,
    YI_val,
    Re_val,
    Dh_val,
    V_val,
    wall_at_left=True
):
    """
    Solve one phase with mixing-length model for a fixed interface flux phi.

    Interface condition remains Neumann:
        mu_eff dU/dY = phi

    No interface velocity is prescribed.
    """
    nut_star = np.zeros_like(Y)
    U_old = None

    for it in range(MIX_MAX_ITER):
        mu_eff = 1.0 / Re_val + nut_star

        if wall_at_left:
            U = solve_ode_fd(
                Y, mu_eff, rhs_const,
                bc_left_type="dirichlet", bc_left_val=0.0,
                bc_right_type="neumann", bc_right_val=phi
            )
        else:
            U = solve_ode_fd(
                Y, mu_eff, rhs_const,
                bc_left_type="neumann", bc_left_val=phi,
                bc_right_type="dirichlet", bc_right_val=0.0
            )

        nut_new = mixing_length_nut_star_from_U(
            Y=Y,
            U=U,
            phase=phase,
            R_val=R_val,
            YI_val=YI_val,
            Re_val=Re_val,
            Dh_val=Dh_val,
            V_val=V_val
        )

        diff_nut = np.linalg.norm(nut_new - nut_star)
        denom_nut = np.linalg.norm(nut_star) + 1e-14

        if U_old is not None:
            diff_U = np.linalg.norm(U - U_old)
            denom_U = np.linalg.norm(U_old) + 1e-14
        else:
            diff_U = np.inf
            denom_U = 1.0

        nut_star = (1.0 - MIX_RELAX) * nut_star + MIX_RELAX * nut_new
        U_old = U.copy()

        if (diff_nut / denom_nut < MIX_TOL) and (diff_U / denom_U < MIX_TOL):
            break

    # Final solve with converged nut
    mu_eff = 1.0 / Re_val + nut_star

    if wall_at_left:
        U = solve_ode_fd(
            Y, mu_eff, rhs_const,
            bc_left_type="dirichlet", bc_left_val=0.0,
            bc_right_type="neumann", bc_right_val=phi
        )
    else:
        U = solve_ode_fd(
            Y, mu_eff, rhs_const,
            bc_left_type="neumann", bc_left_val=phi,
            bc_right_type="dirichlet", bc_right_val=0.0
        )

    nut_star = mixing_length_nut_star_from_U(
        Y=Y,
        U=U,
        phase=phase,
        R_val=R_val,
        YI_val=YI_val,
        Re_val=Re_val,
        Dh_val=Dh_val,
        V_val=V_val
    )

    return U, nut_star


# ============================================================
# CFD POINTS FOR PLOTTING ONLY
# ============================================================

def load_cfd_points(case_id, base_path):
    y_interface = y_I[case_id]

    # Liquid
    file_path_L = os.path.join(base_path, f"Liquid_case_{case_id + 1}.csv")
    data_L = np.loadtxt(file_path_L, delimiter=",", skiprows=1)

    uL_dim = data_L[:, 0]
    yL_dim = data_L[:, 1]

    mask_L = (yL_dim - 0.03) < y_interface

    uL_dim = uL_dim[mask_L]
    yL_dim = yL_dim[mask_L]

    U_Data_L = (uL_dim / Vel_ave_L[case_id])[:, np.newaxis]
    Y_Data_L = ((yL_dim - 0.03) / Dh_L[case_id])[:, np.newaxis]

    # Gas
    file_path_g = os.path.join(base_path, f"Gas_case_{case_id + 1}.csv")
    data_g = np.loadtxt(file_path_g, delimiter=",", skiprows=1)

    uG_dim = data_g[:, 0]
    yG_dim = data_g[:, 1]

    mask_g = (yG_dim - 0.03) > y_interface

    uG_dim = uG_dim[mask_g]
    yG_dim = yG_dim[mask_g]

    U_Data_g = (uG_dim / Vel_ave_g[case_id])[:, np.newaxis]
    Y_Data_g = ((yG_dim - 0.03) / Dh_g[case_id])[:, np.newaxis]

    print(f"Case {case_id + 1}: Liquid CFD points kept = {len(U_Data_L)}")
    print(f"Case {case_id + 1}: Gas CFD points kept    = {len(U_Data_g)}")

    return {
        "uL_plot": U_Data_L[:, 0] * Vel_ave_L[case_id],
        "yL_plot": Y_Data_L[:, 0] * Dh_L[case_id],
        "uG_plot": U_Data_g[:, 0] * Vel_ave_g[case_id],
        "yG_plot": Y_Data_g[:, 0] * Dh_g[case_id],
    }

# ============================================================
# CFD ERROR METRICS
# ============================================================

def compute_velocity_error_against_cfd(r, cfd):
    """
    Compare model velocity against CFD velocity points.

    The model velocity is interpolated at the CFD y-locations.

    Returns errors for:
        liquid phase
        gas phase
        combined liquid+gas
    """

    def phase_error(u_model, y_model, u_cfd, y_cfd):
        # Ensure increasing y for interpolation
        order_model = np.argsort(y_model)
        y_model_sorted = y_model[order_model]
        u_model_sorted = u_model[order_model]

        order_cfd = np.argsort(y_cfd)
        y_cfd_sorted = y_cfd[order_cfd]
        u_cfd_sorted = u_cfd[order_cfd]

        # Interpolate model velocity at CFD y-locations
        u_interp = np.interp(
            y_cfd_sorted,
            y_model_sorted,
            u_model_sorted
        )

        err = u_interp - u_cfd_sorted

        rmse = np.sqrt(np.mean(err**2))
        mae = np.mean(np.abs(err))
        bias = np.mean(err)
        max_abs = np.max(np.abs(err))

        rel_rmse = rmse / max(np.mean(np.abs(u_cfd_sorted)), 1e-14)
        rel_mae = mae / max(np.mean(np.abs(u_cfd_sorted)), 1e-14)

        return {
            "u_interp": u_interp,
            "u_cfd": u_cfd_sorted,
            "y_cfd": y_cfd_sorted,
            "err": err,
            "rmse": rmse,
            "mae": mae,
            "bias": bias,
            "max_abs": max_abs,
            "rel_rmse": rel_rmse,
            "rel_mae": rel_mae,
            "n_points": len(u_cfd_sorted),
        }

    err_L = phase_error(
        u_model=r["uL_phys"],
        y_model=r["yL_phys"],
        u_cfd=cfd["uL_plot"],
        y_cfd=cfd["yL_plot"]
    )

    err_G = phase_error(
        u_model=r["uG_phys"],
        y_model=r["yG_phys"],
        u_cfd=cfd["uG_plot"],
        y_cfd=cfd["yG_plot"]
    )

    # Combined liquid + gas error
    err_all = np.concatenate([err_L["err"], err_G["err"]])
    u_cfd_all = np.concatenate([err_L["u_cfd"], err_G["u_cfd"]])

    rmse_all = np.sqrt(np.mean(err_all**2))
    mae_all = np.mean(np.abs(err_all))
    bias_all = np.mean(err_all)
    max_abs_all = np.max(np.abs(err_all))

    rel_rmse_all = rmse_all / max(np.mean(np.abs(u_cfd_all)), 1e-14)
    rel_mae_all = mae_all / max(np.mean(np.abs(u_cfd_all)), 1e-14)

    return {
        "L": err_L,
        "G": err_G,

        "rmse_all": rmse_all,
        "mae_all": mae_all,
        "bias_all": bias_all,
        "max_abs_all": max_abs_all,
        "rel_rmse_all": rel_rmse_all,
        "rel_mae_all": rel_mae_all,
        "n_points_all": len(u_cfd_all),
    }


# ============================================================
# DISCOVERED-CLOSURE SOLVER
# ============================================================

def solve_case(case_id, N_L_case=N_L, N_G_case=N_G, verbose=False):
    """
    Discovered closure.

    Unknowns:
        phi_L
        phi_g
        log(nut_i_L_star)
        log(nut_i_g_star)

    Enforced:
        meanL = 1
        meanG = 1
        u_i,L = u_i,g
        tau_L = tau_g
    """
    YIL = Y_I_L[case_id]
    YIG = Y_I_g[case_id]

    VL = Vel_ave_L[case_id]
    VG = Vel_ave_g[case_id]

    ReL_val = Re_L[case_id]
    ReG_val = Re_g[case_id]

    dPdXL = dPdX_L[case_id]
    dPdXG = dPdX_g[case_id]

    RL = R_L[case_id]
    RG = R_g[case_id]

    DhL = Dh_L[case_id]
    DhG = Dh_g[case_id]

    nut_i_L_guess = nut_interface_L_star_guess[case_id]
    nut_i_g_guess = nut_interface_g_star_guess[case_id]

    YL = np.linspace(-RL, YIL, N_L_case)
    YG = np.linspace(YIG, RG, N_G_case)

    # Initial liquid solve
    nutL0_star = liquid_nut_star(
        YL, RL, YIL, ReL_val, nut_i_L_guess
    )
    muL0_eff = 1.0 / ReL_val + nutL0_star

    phi_L0, _ = solve_phase_with_mean_constraint(
        YL, muL0_eff, dPdXL, wall_at_left=True
    )

    # Initial gas solve
    gas0 = gas_nut_star(
        YG, RG, YIG, ReL_val, ReG_val, nut_i_g_guess
    )
    nutG0_star = gas0["nut_star"]
    muG0_eff = 1.0 / ReG_val + nutG0_star

    phi_g0, _ = solve_phase_with_mean_constraint(
        YG, muG0_eff, dPdXG, wall_at_left=False
    )

    chi0 = np.array([
        phi_L0,
        phi_g0,
        np.log(max(nut_i_L_guess, 1e-12)),
        np.log(max(nut_i_g_guess, 1e-12)),
    ])

    vel_scale = fixed_velocity_scale(VL, VG)
    shear_scale = fixed_shear_scale(VL, VG)

    def residuals(chi):
        phi_L, phi_g, log_nut_i_L, log_nut_i_g = chi

        nut_i_L_star = np.exp(log_nut_i_L)
        nut_i_g_star = np.exp(log_nut_i_g)

        # Liquid
        nutL_star = liquid_nut_star(
            YL, RL, YIL, ReL_val, nut_i_L_star
        )
        muL_eff = 1.0 / ReL_val + nutL_star

        UL = solve_ode_fd(
            YL, muL_eff, dPdXL,
            bc_left_type="dirichlet", bc_left_val=0.0,
            bc_right_type="neumann", bc_right_val=phi_L
        )

        # Gas
        gas = gas_nut_star(
            YG, RG, YIG, ReL_val, ReG_val, nut_i_g_star
        )
        nutG_star = gas["nut_star"]
        muG_eff = 1.0 / ReG_val + nutG_star

        UG = solve_ode_fd(
            YG, muG_eff, dPdXG,
            bc_left_type="neumann", bc_left_val=phi_g,
            bc_right_type="dirichlet", bc_right_val=0.0
        )

        # Mean constraints
        r_mean_L = trapz_mean(UL, YL) - 1.0
        r_mean_g = trapz_mean(UG, YG) - 1.0

        # Interface velocity continuity
        u_i_L = UL[-1] * VL
        u_i_g = UG[0] * VG
        r_velocity_interface = (u_i_L - u_i_g) / vel_scale

        # Interface shear continuity
        dUL_i = dUdy_last(UL, YL)
        dUG_i = dUdy_first(UG, YG)

        tauL = (1.0 / ReL_val + nut_i_L_star) * dUL_i * rho_L * VL**2
        tauG = (1.0 / ReG_val + nut_i_g_star) * dUG_i * rho_g * VG**2

        r_shear_interface = interface_shear_residual(tauL, tauG) / shear_scale

        return np.array([
            r_mean_L,
            r_mean_g,
            r_velocity_interface,
            r_shear_interface
        ])

    sol = root(residuals, chi0, method="hybr")

    if not sol.success:
        raise RuntimeError(
            f"Case {case_id + 1}: discovered-closure nonlinear solve failed: {sol.message}"
        )

    phi_L_star, phi_g_star, log_nut_i_L, log_nut_i_g = sol.x

    nut_i_L_star = np.exp(log_nut_i_L)
    nut_i_g_star = np.exp(log_nut_i_g)

    # Final liquid
    nutL_star = liquid_nut_star(
        YL, RL, YIL, ReL_val, nut_i_L_star
    )
    muL_eff = 1.0 / ReL_val + nutL_star

    UL = solve_ode_fd(
        YL, muL_eff, dPdXL,
        bc_left_type="dirichlet", bc_left_val=0.0,
        bc_right_type="neumann", bc_right_val=phi_L_star
    )

    # Final gas
    gas = gas_nut_star(
        YG, RG, YIG, ReL_val, ReG_val, nut_i_g_star
    )
    nutG_star = gas["nut_star"]
    muG_eff = 1.0 / ReG_val + nutG_star

    UG = solve_ode_fd(
        YG, muG_eff, dPdXG,
        bc_left_type="neumann", bc_left_val=phi_g_star,
        bc_right_type="dirichlet", bc_right_val=0.0
    )

    mom_res_L = momentum_residual_fd(
        Y=YL,
        U=UL,
        mu_eff=muL_eff,
        rhs_const=dPdXL
    )

    mom_res_g = momentum_residual_fd(
        Y=YG,
        U=UG,
        mu_eff=muG_eff,
        rhs_const=dPdXG
    )

    # Dimensional outputs
    yL_phys = YL * DhL
    yG_phys = YG * DhG

    uL_phys = UL * VL
    uG_phys = UG * VG

    nutL_phys = nutL_star * VL * DhL
    nutG_phys = nutG_star * VG * DhG

    # Interface diagnostics
    dUL_i = dUdy_last(UL, YL)
    dUG_i = dUdy_first(UG, YG)

    tauL = (1.0 / ReL_val + nut_i_L_star) * dUL_i * rho_L * VL**2
    tauG = (1.0 / ReG_val + nut_i_g_star) * dUG_i * rho_g * VG**2

    tau_residual_dim = interface_shear_residual(tauL, tauG)
    final_residuals = residuals(sol.x)

    if verbose:
        print(f"\nCase {case_id + 1}")
        print("Solver success:", sol.success)
        print("Final nonlinear residuals:", final_residuals)
        print("phi_L_star:", phi_L_star)
        print("phi_g_star:", phi_g_star)
        print("nut_i_L_star:", nut_i_L_star)
        print("nut_i_g_star:", nut_i_g_star)

    return {
        "case_id": case_id + 1,
        "model": "Discovered closure",

        "YL": YL,
        "YG": YG,
        "yL_phys": yL_phys,
        "yG_phys": yG_phys,

        "UL": UL,
        "UG": UG,
        "uL_phys": uL_phys,
        "uG_phys": uG_phys,

        "nutL_star": nutL_star,
        "nutG_star": nutG_star,
        "nutL_phys": nutL_phys,
        "nutG_phys": nutG_phys,

        "phi_L_star": phi_L_star,
        "phi_g_star": phi_g_star,
        "nut_i_L_star": nut_i_L_star,
        "nut_i_g_star": nut_i_g_star,

        "gas_nut_peak_star": gas["nut_peak_star"],
        "gas_r_i": gas["r_i"],
        "gas_eta_peak": gas["eta_peak"],
        "gas_c1": gas["c1"],
        "gas_c2": gas["c2"],
        "gas_X": gas["X"],

        "meanL": trapz_mean(UL, YL),
        "meanG": trapz_mean(UG, YG),

        "u_i_L": uL_phys[-1],
        "u_i_g": uG_phys[0],
        "u_i_diff": uL_phys[-1] - uG_phys[0],

        "tauL": tauL,
        "tauG": tauG,
        "tau_residual_dim": tau_residual_dim,

        "final_residuals": final_residuals,

        "mom_res_L_full": mom_res_L["residual_full"],
        "mom_res_g_full": mom_res_g["residual_full"],

        "mom_res_L_rms_abs": mom_res_L["rms_abs"],
        "mom_res_g_rms_abs": mom_res_g["rms_abs"],

        "mom_res_L_rms_rel": mom_res_L["rms_rel"],
        "mom_res_g_rms_rel": mom_res_g["rms_rel"],

        "mom_res_L_max_abs": mom_res_L["max_abs"],
        "mom_res_g_max_abs": mom_res_g["max_abs"],

        "mom_res_L_max_rel": mom_res_L["max_rel"],
        "mom_res_g_max_rel": mom_res_g["max_rel"],

        "root_nfev": sol.nfev,
        "root_success": sol.success,
        "root_message": sol.message,
    }


# ============================================================
# INDEPENDENT MIXING-LENGTH SOLVER
# ============================================================

def solve_case_mixing_length(case_id, N_L_case=N_L, N_G_case=N_G, verbose=False):
    """
    Mixing-length baseline.

    Unknowns:
        phi_L
        phi_g

    Boundary conditions:
        Liquid:
            wall:      U = 0
            interface: mu_eff dU/dY = phi_L

        Gas:
            interface: mu_eff dU/dY = phi_g
            wall:      U = 0

    Enforced:
        u_i,L = u_i,g
        tau_L = tau_g

    Not enforced:
        meanL = 1
        meanG = 1
    """
    YIL = Y_I_L[case_id]
    YIG = Y_I_g[case_id]

    VL = Vel_ave_L[case_id]
    VG = Vel_ave_g[case_id]

    ReL_val = Re_L[case_id]
    ReG_val = Re_g[case_id]

    dPdXL = dPdX_L[case_id]
    dPdXG = dPdX_g[case_id]

    RL = R_L[case_id]
    RG = R_g[case_id]

    DhL = Dh_L[case_id]
    DhG = Dh_g[case_id]

    YL = np.linspace(-RL, YIL, N_L_case)
    YG = np.linspace(YIG, RG, N_G_case)

    vel_scale = fixed_velocity_scale(VL, VG)
    shear_scale = fixed_shear_scale(VL, VG)

    # --------------------------------------------------------
    # Initial guesses only:
    # zero-eddy-viscosity independent phase solves with mean(U)=1
    # --------------------------------------------------------
    muL0 = np.full_like(YL, 1.0 / ReL_val)
    muG0 = np.full_like(YG, 1.0 / ReG_val)

    phi_L0, _ = solve_phase_with_mean_constraint(
        YL, muL0, dPdXL, wall_at_left=True
    )

    phi_g0, _ = solve_phase_with_mean_constraint(
        YG, muG0, dPdXG, wall_at_left=False
    )

    def solve_given_phi(phi_L, phi_g):
        UL, nutL_star = solve_mixing_length_given_phi(
            Y=YL,
            rhs_const=dPdXL,
            phi=phi_L,
            phase="liquid",
            R_val=RL,
            YI_val=YIL,
            Re_val=ReL_val,
            Dh_val=DhL,
            V_val=VL,
            wall_at_left=True
        )

        UG, nutG_star = solve_mixing_length_given_phi(
            Y=YG,
            rhs_const=dPdXG,
            phi=phi_g,
            phase="gas",
            R_val=RG,
            YI_val=YIG,
            Re_val=ReG_val,
            Dh_val=DhG,
            V_val=VG,
            wall_at_left=False
        )

        return UL, UG, nutL_star, nutG_star

    def residuals(chi):
        phi_L, phi_g = chi

        UL, UG, nutL_star, nutG_star = solve_given_phi(phi_L, phi_g)

        # Interface velocity continuity
        u_i_L = UL[-1] * VL
        u_i_g = UG[0] * VG

        r_velocity_interface = (u_i_L - u_i_g) / vel_scale

        # Interface shear continuity
        dUL_i = dUdy_last(UL, YL)
        dUG_i = dUdy_first(UG, YG)

        tauL = (1.0 / ReL_val + nutL_star[-1]) * dUL_i * rho_L * VL**2
        tauG = (1.0 / ReG_val + nutG_star[0]) * dUG_i * rho_g * VG**2

        r_shear_interface = interface_shear_residual(tauL, tauG) / shear_scale

        return np.array([
            r_velocity_interface,
            r_shear_interface
        ])

    chi0 = np.array([phi_L0, phi_g0])

    sol = root(residuals, chi0, method="hybr")

    if not sol.success:
        # Fallback around zero flux
        sol = root(
            residuals,
            np.array([0.0, 0.0]),
            method="hybr"
        )

    if not sol.success:
        raise RuntimeError(
            f"Case {case_id + 1}: mixing-length nonlinear solve failed: {sol.message}"
        )

    phi_L_star, phi_g_star = sol.x

    UL, UG, nutL_star, nutG_star = solve_given_phi(phi_L_star, phi_g_star)

    yL_phys = YL * DhL
    yG_phys = YG * DhG

    uL_phys = UL * VL
    uG_phys = UG * VG

    nutL_phys = nutL_star * VL * DhL
    nutG_phys = nutG_star * VG * DhG

    dUL_i = dUdy_last(UL, YL)
    dUG_i = dUdy_first(UG, YG)

    tauL = (1.0 / ReL_val + nutL_star[-1]) * dUL_i * rho_L * VL**2
    tauG = (1.0 / ReG_val + nutG_star[0]) * dUG_i * rho_g * VG**2

    tau_residual_dim = interface_shear_residual(tauL, tauG)
    final_residuals = residuals(sol.x)

    if verbose:
        print(f"\nCase {case_id + 1} - Mixing length")
        print("Solver success:", sol.success)
        print("Final residuals:", final_residuals)
        print("phi_L_star:", phi_L_star)
        print("phi_g_star:", phi_g_star)
        print("meanL:", trapz_mean(UL, YL))
        print("meanG:", trapz_mean(UG, YG))

    return {
        "case_id": case_id + 1,
        "model": "Mixing length",

        "YL": YL,
        "YG": YG,
        "yL_phys": yL_phys,
        "yG_phys": yG_phys,

        "UL": UL,
        "UG": UG,
        "uL_phys": uL_phys,
        "uG_phys": uG_phys,

        "nutL_star": nutL_star,
        "nutG_star": nutG_star,
        "nutL_phys": nutL_phys,
        "nutG_phys": nutG_phys,

        "phi_L_star": phi_L_star,
        "phi_g_star": phi_g_star,

        "meanL": trapz_mean(UL, YL),
        "meanG": trapz_mean(UG, YG),

        "u_i_L": uL_phys[-1],
        "u_i_g": uG_phys[0],
        "u_i_diff": uL_phys[-1] - uG_phys[0],

        "tauL": tauL,
        "tauG": tauG,
        "tau_residual_dim": tau_residual_dim,

        "final_residuals": final_residuals,

        "root_nfev": sol.nfev,
        "root_success": sol.success,
        "root_message": sol.message,
    }


# ============================================================
# RUN ALL MODELS
# ============================================================

results_closure = [
    solve_case(i, N_L, N_G, verbose=False)
    for i in range(len(Y_I_L))
]

results_mixing = [
    solve_case_mixing_length(i, N_L, N_G, verbose=False)
    for i in range(len(Y_I_L))
]

# Backward-compatible name
results = results_closure


# ============================================================
# SUMMARY TABLE
# ============================================================

# ============================================================
# SUMMARY TABLE
# ============================================================

if PRINT_SUMMARY:
    rows = []

    for i in range(len(Y_I_L)):

        # Load CFD points once per case
        try:
            cfd = load_cfd_points(i, base_path)
        except Exception as e:
            print(f"Case {i + 1}: could not load CFD points for error computation -> {e}")
            cfd = None

        for r in [
            results_closure[i],
            results_mixing[i],
        ]:
            if r["model"] == "Discovered closure":
                res_u = r["final_residuals"][2]
                res_tau = r["final_residuals"][3]
                res_meanL = r["final_residuals"][0]
                res_meanG = r["final_residuals"][1]
            else:
                res_u = r["final_residuals"][0]
                res_tau = r["final_residuals"][1]
                res_meanL = np.nan
                res_meanG = np.nan

            row = {
                "case": r["case_id"],
                "model": r["model"],

                "meanL": r["meanL"],
                "meanG": r["meanG"],
                "meanL_err": r["meanL"] - 1.0,
                "meanG_err": r["meanG"] - 1.0,

                "u_i_L [m/s]": r["u_i_L"],
                "u_i_g [m/s]": r["u_i_g"],
                "u_i_diff [m/s]": r["u_i_diff"],

                "tauL [Pa]": r["tauL"],
                "tauG [Pa]": r["tauG"],
                "tau_residual [Pa]": r["tau_residual_dim"],

                "phi_L_star": r["phi_L_star"],
                "phi_g_star": r["phi_g_star"],

                "res_meanL": res_meanL,
                "res_meanG": res_meanG,
                "res_u_interface": res_u,
                "res_tau_interface": res_tau,

                "nfev": r["root_nfev"],
            }

            # ------------------------------------------------
            # CFD velocity error metrics
            # ------------------------------------------------
            if cfd is not None:
                err = compute_velocity_error_against_cfd(r, cfd)

                row.update({
                    "u_RMSE_L_vs_CFD [m/s]": err["L"]["rmse"],
                    "u_MAE_L_vs_CFD [m/s]": err["L"]["mae"],
                    "u_bias_L_vs_CFD [m/s]": err["L"]["bias"],
                    "u_maxabs_L_vs_CFD [m/s]": err["L"]["max_abs"],
                    "u_relRMSE_L_vs_CFD": err["L"]["rel_rmse"],

                    "u_RMSE_G_vs_CFD [m/s]": err["G"]["rmse"],
                    "u_MAE_G_vs_CFD [m/s]": err["G"]["mae"],
                    "u_bias_G_vs_CFD [m/s]": err["G"]["bias"],
                    "u_maxabs_G_vs_CFD [m/s]": err["G"]["max_abs"],
                    "u_relRMSE_G_vs_CFD": err["G"]["rel_rmse"],

                    "u_RMSE_all_vs_CFD [m/s]": err["rmse_all"],
                    "u_MAE_all_vs_CFD [m/s]": err["mae_all"],
                    "u_bias_all_vs_CFD [m/s]": err["bias_all"],
                    "u_maxabs_all_vs_CFD [m/s]": err["max_abs_all"],
                    "u_relRMSE_all_vs_CFD": err["rel_rmse_all"],
                    "CFD_points_all": err["n_points_all"],
                })

                # Store inside result dictionary too, useful for plotting
                r["cfd_velocity_error"] = err

            else:
                row.update({
                    "u_RMSE_L_vs_CFD [m/s]": np.nan,
                    "u_MAE_L_vs_CFD [m/s]": np.nan,
                    "u_bias_L_vs_CFD [m/s]": np.nan,
                    "u_maxabs_L_vs_CFD [m/s]": np.nan,
                    "u_relRMSE_L_vs_CFD": np.nan,

                    "u_RMSE_G_vs_CFD [m/s]": np.nan,
                    "u_MAE_G_vs_CFD [m/s]": np.nan,
                    "u_bias_G_vs_CFD [m/s]": np.nan,
                    "u_maxabs_G_vs_CFD [m/s]": np.nan,
                    "u_relRMSE_G_vs_CFD": np.nan,

                    "u_RMSE_all_vs_CFD [m/s]": np.nan,
                    "u_MAE_all_vs_CFD [m/s]": np.nan,
                    "u_bias_all_vs_CFD [m/s]": np.nan,
                    "u_maxabs_all_vs_CFD [m/s]": np.nan,
                    "u_relRMSE_all_vs_CFD": np.nan,
                    "CFD_points_all": np.nan,
                })

            if r["model"] == "Discovered closure":
                row.update({
                    "nut_i_L_star": r["nut_i_L_star"],
                    "nut_i_g_star": r["nut_i_g_star"],

                    "momL_rms_rel": r["mom_res_L_rms_rel"],
                    "momG_rms_rel": r["mom_res_g_rms_rel"],

                    "gas_r_i": r["gas_r_i"],
                    "gas_eta_peak": r["gas_eta_peak"],
                    "gas_c1": r["gas_c1"],
                    "gas_c2": r["gas_c2"],
                })
            else:
                row.update({
                    "nut_i_L_star": np.nan,
                    "nut_i_g_star": np.nan,

                    "momL_rms_rel": np.nan,
                    "momG_rms_rel": np.nan,

                    "gas_r_i": np.nan,
                    "gas_eta_peak": np.nan,
                    "gas_c1": np.nan,
                    "gas_c2": np.nan,
                })

            rows.append(row)

    df_summary_all = pd.DataFrame(rows)

    print("\n================ FD VALIDATION SUMMARY: CLOSURE + MIXING LENGTH + CFD ERROR ================\n")
    print(df_summary_all.to_string(index=False))

    print("\n================ COMPACT CFD VELOCITY ERROR TABLE ================\n")
    error_cols = [
        "case",
        "model",
        "u_RMSE_L_vs_CFD [m/s]",
        "u_RMSE_G_vs_CFD [m/s]",
        "u_RMSE_all_vs_CFD [m/s]",
        "u_MAE_all_vs_CFD [m/s]",
        "u_bias_all_vs_CFD [m/s]",
        "u_relRMSE_all_vs_CFD",
        "CFD_points_all",
    ]

    print(df_summary_all[error_cols].to_string(index=False))


# ============================================================
# OPTIONAL GRID-REFINEMENT CHECK FOR DISCOVERED CLOSURE ONLY
# ============================================================

def grid_refinement_check(case_id, grid_sizes=(200, 400, 800)):
    rows = []

    for N in grid_sizes:
        r = solve_case(case_id, N_L_case=N, N_G_case=N, verbose=False)

        rows.append({
            "case": case_id + 1,
            "N_L": N,
            "N_G": N,

            "meanL_err": r["meanL"] - 1.0,
            "meanG_err": r["meanG"] - 1.0,

            "momL_rms_rel": r["mom_res_L_rms_rel"],
            "momG_rms_rel": r["mom_res_g_rms_rel"],

            "u_i_diff [m/s]": r["u_i_diff"],
            "tau_residual [Pa]": r["tau_residual_dim"],

            "nut_i_L_star": r["nut_i_L_star"],
            "nut_i_g_star": r["nut_i_g_star"],

            "nfev": r["root_nfev"],
        })

    return pd.DataFrame(rows)


if RUN_GRID_REFINEMENT:
    for case_id in range(len(Y_I_L)):
        print(f"\nGrid-refinement check: Case {case_id + 1}")
        print(grid_refinement_check(case_id).to_string(index=False))


# ============================================================
# COMBINED PLOTS
# ============================================================


COLOR_CLOSURE = "tab:blue"
COLOR_MIXING  = "tab:orange"

STYLE_LIQUID = "-"
STYLE_GAS    = "-"

if PLOT_RESULTS:
    for i in range(len(Y_I_L)):
        r_cl = results_closure[i]
        r_ml = results_mixing[i]

        case_id = i

        fig, ax = plt.subplots(1, 2, figsize=(13, 5))

        cfd = None

        if PLOT_CFD_DOTS:
            try:
                cfd = load_cfd_points(case_id, base_path)
            except Exception as e:
                print(f"Case {case_id + 1}: could not load CFD dots -> {e}")

        # ----------------------------------------------------
        # Velocity plot
        # ----------------------------------------------------

        ax[0].plot(
            r_cl["uL_phys"], r_cl["yL_phys"],
            color=COLOR_CLOSURE,
            linestyle=STYLE_LIQUID,
            linewidth=1.5,
            label="Closure"
        )

        ax[0].plot(
            r_cl["uG_phys"], r_cl["yG_phys"],
            color=COLOR_CLOSURE,
            linestyle=STYLE_GAS,
            linewidth=1.5
        )

        ax[0].plot(
            r_ml["uL_phys"], r_ml["yL_phys"],
            color=COLOR_MIXING,
            linestyle=STYLE_LIQUID,
            linewidth=1.5,
            label="Mixing length"
        )

        ax[0].plot(
            r_ml["uG_phys"], r_ml["yG_phys"],
            color=COLOR_MIXING,
            linestyle=STYLE_GAS,
            linewidth=1.5
        )

        if cfd is not None:
            ax[0].plot(
                cfd["uL_plot"], cfd["yL_plot"],
                ".",
                color="black",
                markersize=4,
                label="CFD"
            )

            ax[0].plot(
                cfd["uG_plot"], cfd["yG_plot"],
                ".",
                color="black",
                markersize=4
            )

        ax[0].axhline(
            y_I[case_id],
            linestyle="--",
            color="k",
            alpha=0.5,
            label="Interface"
        )

        ax[0].set_xlabel("u [m/s]")
        ax[0].set_ylabel("y [m]")
        ax[0].set_title(f"Case {case_id + 1}: velocity comparison")
        ax[0].grid(True)
        ax[0].legend(fontsize=8)

        # ----------------------------------------------------
        # Eddy-viscosity plot
        # ----------------------------------------------------

        ax[1].plot(
            r_cl["nutL_phys"], r_cl["yL_phys"],
            color=COLOR_CLOSURE,
            linestyle=STYLE_LIQUID,
            linewidth=1.5,
            label="Closure"
        )

        ax[1].plot(
            r_cl["nutG_phys"], r_cl["yG_phys"],
            color=COLOR_CLOSURE,
            linestyle=STYLE_GAS,
            linewidth=1.5
        )

        ax[1].plot(
            r_ml["nutL_phys"], r_ml["yL_phys"],
            color=COLOR_MIXING,
            linestyle=STYLE_LIQUID,
            linewidth=1.5,
            label="Mixing length"
        )

        ax[1].plot(
            r_ml["nutG_phys"], r_ml["yG_phys"],
            color=COLOR_MIXING,
            linestyle=STYLE_GAS,
            linewidth=1.5
        )

        ax[1].axhline(
            y_I[case_id],
            linestyle="--",
            color="k",
            alpha=0.5,
            label="Interface"
        )

        ax[1].set_xlabel(r"$\nu_t$ [m$^2$/s]")
        ax[1].set_ylabel("y [m]")
        ax[1].set_title(f"Case {case_id + 1}: eddy-viscosity comparison")
        ax[1].grid(True)
        ax[1].legend(fontsize=8)

        plt.tight_layout()
        plt.show()


# ============================================================
# OPTIONAL MOMENTUM-RESIDUAL PLOTS FOR DISCOVERED CLOSURE ONLY
# ============================================================

if PLOT_MOMENTUM_RESIDUALS:
    for r in results_closure:
        case_id = r["case_id"] - 1

        fig_res, ax_res = plt.subplots(1, 2, figsize=(12, 4))

        ax_res[0].plot(
            r["mom_res_L_full"],
            r["yL_phys"],
            label="Liquid momentum residual"
        )

        ax_res[0].axhline(
            y_I[case_id],
            linestyle="--",
            color="k",
            alpha=0.5,
            label="Interface"
        )

        ax_res[0].set_xlabel(r"$R_{mom,L}$")
        ax_res[0].set_ylabel("y [m]")
        ax_res[0].set_title(
            f"Case {case_id + 1}: liquid momentum residual"
        )
        ax_res[0].grid(True)
        ax_res[0].legend()

        ax_res[1].plot(
            r["mom_res_g_full"],
            r["yG_phys"],
            label="Gas momentum residual"
        )

        ax_res[1].axhline(
            y_I[case_id],
            linestyle="--",
            color="k",
            alpha=0.5,
            label="Interface"
        )

        ax_res[1].set_xlabel(r"$R_{mom,g}$")
        ax_res[1].set_ylabel("y [m]")
        ax_res[1].set_title(
            f"Case {case_id + 1}: gas momentum residual"
        )
        ax_res[1].grid(True)
        ax_res[1].legend()

        plt.tight_layout()
        plt.show()
```
